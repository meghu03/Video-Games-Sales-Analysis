```python
#importing necessary libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
sheet_name = 'videogames_sales'
sheet_id = '1t6kVv95bVRQ9NyC1h3pO06WH8tREZDGq'
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
```

```python
#Loading the dataset
data = pd.read_csv(url)
data
```

```python
#Removing rows containing null values
df = pd.DataFrame(data)
df.dropna(inplace=True)
df
```

```python
#Checking for duplicates
df = df.drop_duplicates()
df.duplicated().values.any()
```

```python
#Checking for Null values 
pd.isnull(df).any()
```

```python
#Renaming the columns
data = data.rename(columns={'Name':'Games Name','Platform':'Platforms'})
data
```

```python
import pandas as pd
from matplotlib import pyplot as plt
data = df
data.head()
df = pd.DataFrame(data)
plt.figure(figsize = (10,10))
plt.bar(df["Year"].value_counts().index, df["Year"].value_counts())
plt.xlabel("Years")
plt.ylabel("Count of games")
plt.title("Number of releases of video games for Years")
plt.show()
```
